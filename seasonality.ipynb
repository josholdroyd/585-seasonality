```python
import numpy as np
import polars as pl
import matplotlib.pyplot as plt
import datetime as dt
```

```python
df = pl.scan_ipc('crsp_daily.ftr').collect()

df
```

```python
# add log return col and remove unnecessary cols

df = (
    df
    .with_columns(
        pl.col('caldt').dt.date()
    )
    .with_columns(
        pl.col('ret').log1p().alias('logret')
    )
    .drop(['siccd','vol','shr','excd'])
)

df
```

```python
# filter to dates from paper

df = df.filter((pl.col('caldt') >= dt.date(1965,1,1)) & (pl.col('caldt') <= dt.date(2002,12,31)))

df
```

```python
# create ret_21d monthly returns on daily basis for extension

df = (
    df
    .sort(['permno','caldt'])
    .with_columns(
        pl.col('logret').rolling_sum(window_size=21).over('permno').exp().sub(1).alias('ret_21d')
    )
)

# create ret_1m monthly returns for replication

df = (
    df
    .with_columns(
        pl.col('caldt').dt.truncate('1mo').alias('month')
    )
    .with_columns(
        pl.col('logret').sum().over(['permno','month']).exp().sub(1).alias('ret_1m')
    )
    .sort(['permno','month'])
)

df
```

```python
# Step 1: create a monthly-level DataFrame
monthly = (
    df.select(["permno", "month", "ret_1m"])
      .unique(subset=["permno", "month"])
      .sort(["permno", "month"])
)

# Step 2: shift by k months
k_vals = [1, 2, 3, 6, 12]

for k in k_vals:
    monthly = monthly.with_columns(
        pl.col("ret_1m")
          .shift(k)  # shift by k months
          .over("permno")
          .alias(f"ret_1m_lag{k}m")
    )

# Step 3: merge back to daily-level DataFrame
df = df.join(
    monthly.select(["permno", "month"] + [f"ret_1m_lag{k}m" for k in k_vals]),
    on=["permno", "month"],
    how="left"
)

df.head(20)
```

```python
import statsmodels.api as sm

results = {}

for k in k_vals:
    X = df[f"ret_1m_lag{k}m"].to_numpy()  # predictor
    y = df["ret_1m"].to_numpy()           # dependent variable
    
    # Drop rows with missing lag
    mask = ~np.isnan(X) & ~np.isnan(y)
    X = X[mask]
    y = y[mask]
    
    # Add constant for intercept
    X = sm.add_constant(X)
    
    # Fit OLS
    model = sm.OLS(y, X).fit()
    
    results[k] = model
    print(f"\nRegression for lag {k} months")
    print(model.summary())
```
